```python
import pandas as pd
import openpyxl
```

```python
data = pd.read_excel("Planilha Jr.xlsx", index_col=None)
data.head()
```

```python
data.rename(str.title, inplace=True, axis=1)
data.rename({"Aluno": "Nome", "Apto":"Apartamento", "Aloj":"Bloco"}, inplace=True, axis=1)
data.head()
```

```python
data.dropna(axis=0, inplace=True, subset=["Bloco", "Apartamento","Vaga", "Nome", "Email"])
data.reset_index(col_level=0, inplace=True)
data.rename({"index":"ID"}, inplace=True, axis=1)
data.head()
```

```python
data["Apartamento"] = data["Apartamento"].apply(lambda x:int(x))
data["Bloco"] = data["Bloco"].apply(lambda x:x[-1])
data.head()
```

```python
data.to_csv("moradores.xlsx", index=False, sheet_name="Entregas", engine="openpyxl")
```
